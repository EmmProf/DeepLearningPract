```python
transactions.date = pd.to_datetime(transactions.date,format="%d.%m.%Y")
```

```python
transactionsPlus = transactions.merge(items,
                                      on = "item_id",
                                      sort=False).merge(items_categ,
                                                        on="item_category_id",
                                                        sort=False).merge(shops,
                                                                          on = 'shop_id',
                                                                          sort=False)
```

```python
transactionsPlus.shop_id = transactionsPlus.shop_id.astype('str')

transactionsPlus.item_id = transactionsPlus.item_id.astype('str')

transactionsPlus['shopItemID'] =transactionsPlus[['shop_id','item_id']].apply(lambda s: ",".join(s),axis=1)
```
